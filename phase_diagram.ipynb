```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
```

```python
file_path_SIM = fr'CO2_Phase_diagram1.txt'
SIM1 = np.loadtxt(file_path_SIM, delimiter=',')
file_path_SIM = fr'CO2_Phase_diagram2.txt'
SIM2 = np.loadtxt(file_path_SIM, delimiter=',')
file_path_SIM = fr'CO2_Phase_diagram3.txt'
SIM3 = np.loadtxt(file_path_SIM, delimiter=',')
```

```python
with plt.style.context(['ieee']):
    plt.rcParams['font.serif'] = ['Times New Roman']
    fig, ax = plt.subplots()

    # Plot the data with a log scale on the x-axis
    ax.plot(SIM1[:, 0], SIM1[:, 1], linestyle='-', linewidth=0.75, color='#4daf4a', label='SIM1')
    ax.plot(SIM2[:, 0], SIM2[:, 1], linestyle='-', linewidth=0.75, color='#e41a1c', label='SIM2')
    ax.plot(SIM3[:, 0], SIM3[:, 1], linestyle='-', linewidth=0.75, color='#377eb8', label='SIM3')

    # Add vertical and horizontal dashed lines with hex code color
    #plt.axvline(x=304.128, color='k', linestyle='--', linewidth=0.5)
    #plt.axhline(y=7.3773, color='k', linestyle='--', linewidth=0.5)
    
    ax.plot([304.128, 400], [7.3773, 7.3773], color='k', linestyle='--', linewidth=0.5)
    ax.plot([304.128, 304.128], [7.3773, 576.5932649858933], color='k', linestyle='--', linewidth=0.5)
    
    point1 = (304.128, 7.3773)
    point2 = (216.4926931106472, 0.5216645251859957)

    plt.scatter(*point1, color='k', s=5, label='Point 1')
    plt.scatter(*point2, color='k', s=5, label='Point 2')

    ax.set_yscale('log')
    ax.set_xlabel('Temperature / [K]')
    ax.set_ylabel('Pressure / [MPa]')

    ax.tick_params(axis='both', which='both', direction='in')
    ax.set_xlim(200, 400)
    ax.set_ylim(0, 1000)
    
    ax.text(230, 2, 'Boiling curve', fontsize=6, color='#e41a1c')
    ax.text(305, 10, 'Critical point', fontsize=6, color='k')
    ax.text(275,0.5, 'Gas', fontsize=6, color='k')
    ax.text(350,80, 'Supercritical \n fluid', fontsize=6, color='k', ha='center', va='center')
    ax.text(275,80, 'Liquid', fontsize=6, color='k', ha='center', va='center')
    fig.savefig(r'phase_diagram_K.jpg', dpi=1200,  bbox_inches='tight')

plt.show()
```

```python
with plt.style.context(['ieee']):
    plt.rcParams['font.serif'] = ['Times New Roman']
    fig, ax = plt.subplots()

    # Plot the data with a log scale on the x-axis
    ax.plot(SIM1[:, 0]-273, SIM1[:, 1]*10, linestyle='-', linewidth=0.75, color='#4daf4a', label='SIM1')
    ax.plot(SIM2[:, 0]-273, SIM2[:, 1]*10, linestyle='-', linewidth=0.75, color='#e41a1c', label='SIM2')
    ax.plot(SIM3[:, 0]-273, SIM3[:, 1]*10, linestyle='-', linewidth=0.75, color='#377eb8', label='SIM3')

    # Add vertical and horizontal dashed lines with hex code color
    #plt.axvline(x=304.128, color='k', linestyle='--', linewidth=0.5)
    #plt.axhline(y=7.3773, color='k', linestyle='--', linewidth=0.5)
    
    ax.plot([304.128-273, 80], [7.3773*10, 7.3773*10], color='k', linestyle='--', linewidth=0.5)
    ax.plot([304.128-273, 304.128-273], [7.3773*10, 576.5932649858933*10], color='k', linestyle='--', linewidth=0.5)
    
    ax.text(230-273, 2*10, 'Boiling curve', fontsize=6, color='#e41a1c')
    ax.text(305-273, 10*10, 'Critical point', fontsize=6, color='k')
    ax.text(275-273,0.5*10, 'Gas', fontsize=6, color='k')
    ax.text(55,80*10, 'Supercritical \n fluid', fontsize=6, color='k', ha='center', va='center')
    ax.text(275-273,80*10, 'Liquid', fontsize=6, color='k', ha='center', va='center')

    point1 = (304.128-273, 7.3773*10)
    point2 = (216.4926931106472-273, 0.5216645251859957*10)
    plt.scatter(*point1, color='k', s=5, label='Point 1')
    plt.scatter(*point2, color='k', s=5, label='Point 2')

    ax.set_yscale('log')
    ax.grid(which='both')

    ax.grid(which='minor', linewidth=0.25, color='black', linestyle='--', alpha=0.2)
    ax.grid(which='major', linewidth=0.5, color='black', linestyle='-', alpha=0.5)

    ax.set_xlabel('Temperature / [$^\circ$C]')
    ax.set_ylabel('Pressure / [bar]')
    
    x_interval = 20
    x_ticks = np.arange(-80, 80, 20)
    ax.set_xticks(x_ticks)
    ax.set_xticks(np.arange(-80, 100, x_interval), minor=True)
    ax.tick_params(axis='both', which='both', direction='in')
    ax.set_xlim(-73,80)
    ax.set_ylim(0, 10000)
    fig.savefig(r'phase_diagram_C.jpg', dpi=1200,  bbox_inches='tight')
    
plt.show()
```
