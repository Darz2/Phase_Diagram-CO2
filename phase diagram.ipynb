```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Set the axis labels
ax.set_xlabel('Angstrom')
ax.set_ylabel('Time (years)')

# Add a title to the plot
ax.set_title('Hierarchy of computational chemical methods')

# Add a grid to the plot
ax.grid(True)

# Define the text labels for the different levels of the hierarchy
hierarchy_labels = ['Atoms', 'Molecular conformations', 'Bond formation',
                   'Design', 'FEA', 'MD', 'MESO', 'QC', 'ReaxFF']

# Add the text labels to the plot
for i, label in enumerate(hierarchy_labels):
    ax.text(i, 15, label, ha='center', va='center')

# Add arrows to connect the different levels of the hierarchy
for i in range(len(hierarchy_labels) - 1):
    ax.arrow(i, 15 - i, 1, -1, head_width=0.2, head_length=0.3, color='black')

# Set the limits of the plot
ax.set_xlim(-0.5, len(hierarchy_labels) - 0.5)
ax.set_ylim(0, 15)

# Save the plot to a file
fig.savefig('hierarchy_of_computational_chemical_methods.png')
plt.show()
```
